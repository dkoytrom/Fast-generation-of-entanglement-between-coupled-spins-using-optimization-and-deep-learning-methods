```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(2, 2, 1, projection='3d')
ax2 = fig.add_subplot(2, 2, 2, projection='3d')
ax3 = fig.add_subplot(2, 2, 3, projection='3d')

ax1.view_init(40, 10, 0)
ax2.view_init(40, 10, 0)
ax3.view_init(40, 10, 0)

fig.set_figheight(8)
fig.set_figwidth(11)

x = [0, 1, 0, 0] # P1
y = [1, 0, 0, 1] # P3
z = [0, 0, 1, 0] # P2
pop_labels = ["P1 = 1", "P3 = 1", "P2 = 1"]
colors = ['r', 'g', 'b', 'orange']

ax1.plot(x, y, z, zdir='z', linestyle='--')
ax1.scatter(x, y, z, zdir='z', s=200)
ax2.plot(x, y, z, zdir='z', linestyle='--')
ax2.scatter(x, y, z, zdir='z', s=200)
ax3.plot(x, y, z, zdir='z', linestyle='--')
ax3.scatter(x, y, z, zdir='z', s=200)

#pso conc
labels = [r"PSO - $(\Omega,\Delta)_{max} = 1 \xi$", r"PSO - $(\Omega,\Delta)_{max} = 2 \xi$", r"PSO - $(\Omega,\Delta)_{max} = 3 \xi$", r"PSO - $(\Omega,\Delta)_{max} = 4 \xi$"]

p1 = [0.0476, 0.2587, 0.2952, 0.3059] # P1
p3 = [0.0447, 0.2656, 0.2912, 0.3038] # P3
p2 = [0.9077, 0.4756, 0.4136, 0.3902] # P2

for i, txt in enumerate(pop_labels):
    ax1.text(x[i], y[i] + 0.05, z[i], txt)

for i in range(4):
    ax1.scatter(p1[i], p3[i], p2[i], marker='o', s=100, c=colors[i], label=labels[i])

ax1.legend(loc='upper left', bbox_to_anchor=(1.2, -0.2), ncol=1, fancybox=True)
ax1.set_title('(a)', loc = "left", fontsize = 10)

# slsqp conc
labels = [r"SLSQP - $(\Omega,\Delta)_{max} = 1 \xi$", r"SLSQP - $(\Omega,\Delta)_{max} = 2 \xi$", r"SLSQP - $(\Omega,\Delta)_{max} = 3 \xi$", r"SLSQP - $(\Omega,\Delta)_{max} = 4 \xi$"]

p1 = [0.0787, 0.2614, 0.2301, 0.3245] # P1
p3 = [0.0760, 0.2610, 0.2298, 0.3235] # P3
p2 = [0.8453, 0.4776, 0.5401, 0.3520] # P2

for i, txt in enumerate(pop_labels):
    ax2.text(x[i], y[i] + 0.05, z[i], txt)

for i in range(4):
    ax2.scatter(p1[i], p3[i], p2[i], marker='*', s=100, c=colors[i], label=labels[i])

ax2.legend(loc='upper left', bbox_to_anchor=(0.2, -0.2), ncol=1, fancybox=True)
ax2.set_title('(b)', loc = "left", fontsize = 10)

# BBNN conc
labels = [r"BBNN - $(\Omega,\Delta)_{max} = 1 \xi$", r"BBNN - $(\Omega,\Delta)_{max} = 2 \xi$", r"BBNN - $(\Omega,\Delta)_{max} = 3 \xi$", r"BBNN - $(\Omega,\Delta)_{max} = 4 \xi$"]

p1 = [0.0801, 0.4042, 0.3338, 0.2928] # P1
p3 = [0.0861, 0.4112, 0.3157, 0.3532] # P3
p2 = [0.8338, 0.1846, 0.3505, 0.3540] # P2

for i, txt in enumerate(pop_labels):
    ax3.text(x[i], y[i] + 0.05, z[i], txt)

for i in range(4):
    ax3.scatter(p1[i], p3[i], p2[i], marker='x', s=100, c=colors[i], label=labels[i])

ax3.legend(loc='upper left', bbox_to_anchor=(1.5, 0.6), ncol=1, fancybox=True)
ax3.set_title('(c)', loc = "left", fontsize = 10)
```

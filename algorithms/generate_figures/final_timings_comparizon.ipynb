```python
import matplotlib.pyplot as plt
```

```python
final_times_PSO_conc = [2.3, 1.3, 1.2, 1.0]
final_times_PSO_fid = [2.6, 1.4, 1.3, 1.2]
final_times_SLSQP_conc = [2.1, 1.2, 1.1, 1]
final_times_SLSQP_fid = [2.45, 1.4, 1.2, 1.1]
final_times_BBNN_conc = [2.1, 1.3, 1.1, 1]
final_times_BBNN_fid = [2.4, 1.4, 1.2, 1.1]

max_control = [1, 2, 3, 4]
```

```python
fig, ax = plt.subplots(3, 1, sharex='col')
fig.set_figheight(10)
fig.set_figwidth(8)

ax1 = ax[0]
ax2 = ax[1]
ax3 = ax[2]

# PSO results
ax1.plot(max_control, final_times_PSO_conc, label="PSO - Optimizing concurrence")
ax1.scatter(max_control, final_times_PSO_conc, marker='p')
ax1.plot(max_control, final_times_PSO_fid, label="PSO - Optimizing fidelity")
ax1.scatter(max_control, final_times_PSO_fid, marker='x')

ax1.set_ylabel(r"Minimum time T ($\xi^{-1}$)")
ax1.set_xlabel(r"Max control absolute value ($\xi$)")
ax1.legend()
ax1.set_title('(a)', loc = "right", fontsize = 10)

# SLSQP results
ax2.plot(max_control, final_times_SLSQP_conc, label="SLSQP - Optimizing concurrence")
ax2.scatter(max_control, final_times_SLSQP_conc, marker='p')
ax2.plot(max_control, final_times_SLSQP_fid, label="SLSQP - Optimizing fidelity")
ax2.scatter(max_control, final_times_SLSQP_fid, marker='x')

ax2.set_ylabel(r"Minimum time T ($\xi^{-1}$)")
ax2.set_xlabel(r"Max control absolute value ($\xi$)")
ax2.legend()
ax2.set_title('(b)', loc = "right", fontsize = 10)

# BBNN results
ax3.plot(max_control, final_times_BBNN_conc, label="BBNN - Optimizing concurrence")
ax3.scatter(max_control, final_times_BBNN_conc, marker='p')
ax3.plot(max_control, final_times_BBNN_fid, label="BBNN - Optimizing fidelity")
ax3.scatter(max_control, final_times_BBNN_fid, marker='x')

ax3.set_ylabel(r"Minimum time T ($\xi^{-1}$)")
ax3.set_xlabel(r"Max control absolute value ($\xi$)")
ax3.legend()
ax3.set_title('(c)', loc = "right", fontsize = 10)
```

```python
import numpy as np

max_controls = (r"$(\Omega,\Delta)_{max} = 1 \xi$", r"$(\Omega,\Delta)_{max} = 2 \xi$", r"$(\Omega,\Delta)_{max} = 3 \xi$", r"$(\Omega,\Delta)_{max} = 4 \xi$")
minimum_times_C = {
    'PSO': (2.3, 1.3, 1.2, 1.0),
    'SLSQP': (2.1, 1.2, 1.1, 1.0),
    'BBNN': (2.1, 1.3, 1.1, 1.0),
}

minimum_times_F = {
    'PSO': (2.6, 1.4, 1.3, 1.2),
    'SLSQP': (2.45, 1.4, 1.2, 1.1),
    'BBNN': (2.4, 1.4, 1.2, 1.1),
}

x = np.arange(len(max_controls))  # the label locations
width = 0.25  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(1, 2, layout='constrained')
ax1 = ax[0]
ax2 = ax[1]

fig.set_figwidth(10)

for attribute, measurement in minimum_times_C.items():
    offset = width * multiplier
    rects = ax1.bar(x + offset, measurement, width, label=attribute)
    ax1.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax1.set_ylabel(r'Minimum time T ($\xi^{-1}$)')
ax1.set_title(r'Maximizing Concurrence $\mathcal{C}$')
ax1.set_xticks(x + width, max_controls)
ax1.legend(loc='upper right', ncols=3)
ax1.set_ylim(0, 3)
ax1.set_title('(a)', loc = "right", fontsize = 10)

for attribute, measurement in minimum_times_F.items():
    offset = width * multiplier
    rects = ax2.bar(x + offset, measurement, width, label=attribute)
    ax2.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax2.set_ylabel(r'Minimum time T ($\xi^{-1}$)')
ax2.set_title(r'Maximizing Fidelity $\mathcal{F}$')
ax2.set_xticks(x + width + 0.75, max_controls)
ax2.legend(loc='upper right', ncols=3)
ax2.set_ylim(0, 3)
ax2.set_title('(b)', loc = "right", fontsize = 10)

plt.show()
```
